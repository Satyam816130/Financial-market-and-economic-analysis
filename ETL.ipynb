```python
import zipfile
import os

# Define the zip file path
zip_path = "finance-and-economics-dataset-2000-present.zip"

# Extract the files
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("finance-and-economics-dataset-2000-present")  # Extract into a folder

# Check extracted files
print("Extracted files:", os.listdir("finance-and-economics-dataset-2000-present"))
```

```python
import pandas as pd

# Find the CSV file
folder_path = "finance-and-economics-dataset-2000-present"
files = os.listdir(folder_path)
csv_file = [file for file in files if file.endswith(".csv")]

if csv_file:
    # Load the CSV into Pandas
    df = pd.read_csv(os.path.join(folder_path, csv_file[0]))
    print(df.head())  # Display the first few rows
else:
    print("No CSV file found in the extracted folder.")
```
